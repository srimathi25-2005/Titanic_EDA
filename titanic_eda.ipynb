```python
# TITANIC EDA NOTEBOOK TEMPLATE
# Objective: Extract insights using visual and statistical exploration

# -------------------------
# 1. Import Libraries
# -------------------------
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
%matplotlib inline

# -------------------------
# 2. Load Dataset
# -------------------------
df = pd.read_csv(r"C:\Users\srima\Task 1\Titanic\train.csv")  # Replace with your path

# Quick look at data
print(df.head())

# -------------------------
# 3. Basic Exploration
# -------------------------
print(df.info())
print(df.describe(include='all'))
print(df.isnull().sum())

# -------------------------
# 4. Univariate Analysis
# -------------------------
# Categorical variables
sns.countplot(x='Survived', data=df).set_title('Survival Count')
plt.show()
sns.countplot(x='Sex', data=df).set_title('Gender Distribution')
plt.show()
sns.countplot(x='Pclass', data=df).set_title('Passenger Class Distribution')
plt.show()

# Numerical variables
sns.histplot(df['Age'], bins=30, kde=True).set_title('Age Distribution')
plt.show()
sns.boxplot(x=df['Fare']).set_title('Fare Distribution')
plt.show()

# -------------------------
# 5. Bivariate Analysis
# -------------------------
# Survival by Gender & Class
sns.countplot(x='Sex', hue='Survived', data=df).set_title('Survival by Gender')
plt.show()
sns.countplot(x='Pclass', hue='Survived', data=df).set_title('Survival by Passenger Class')
plt.show()

# Age & Fare vs Survival
sns.histplot(data=df, x='Age', hue='Survived', multiple='stack').set_title('Age vs Survival')
plt.show()
sns.boxplot(x='Survived', y='Fare', data=df).set_title('Fare vs Survival')
plt.show()

# -------------------------
# 6. Multivariate Analysis
# -------------------------
numeric_df = df.select_dtypes(include=['int64', 'float64'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass']], hue='Survived')
plt.show()
print(df.dtypes)

```
